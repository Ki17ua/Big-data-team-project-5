```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')

# Load data from CSV files
def load_data(file_path):
    df = pd.read_csv(file_path)
    # Strip any whitespace from column names to prevent issues
    df.columns = df.columns.str.strip()
    return df

# Basic summary statistics
def summarize_data(df):
    return df.describe()

# Corrected Plot data function to dynamically set labels
def plot_data(df, title):
    plt.figure(figsize=(14, 7))  # Adjust size for better visibility
    time_column = df.columns[0]  # Assuming first column is 'Time Interval'
    df[time_column] = df[time_column].astype(str)  # Ensure time column is string for plotting

    # Iterate over each column to plot, skipping 'Time Interval'
    for column in df.columns[1:]:
        plt.plot(df[time_column], df[column], marker='o', label=column)  # Label each line by column name

    plt.title(title)
    plt.xlabel('Time Interval')
    plt.ylabel('Counts')
    plt.xticks(rotation=45)
    plt.legend(loc='upper left')  # Place legend to avoid obscuring the plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# File paths
cycle_path = '/content/drive/MyDrive/cycle.csv'
pedestrian_path = '/content/drive/MyDrive/pedestrian.csv'
old_pedestrian_path = '/content/drive/MyDrive/old pedestrain.csv'
vehicle_path = '/content/drive/MyDrive/vehicle.csv'

# Load datasets
cycle_data = load_data(cycle_path)
pedestrian_data = load_data(pedestrian_path)
old_pedestrian_data = load_data(old_pedestrian_path)
vehicle_data = load_data(vehicle_path)

# Summarize data
print("Cycle Data Summary:")
print(summarize_data(cycle_data))
print("Pedestrian Data Summary:")
print(summarize_data(pedestrian_data))
print("Old Pedestrian Data Summary:")
print(summarize_data(old_pedestrian_data))
print("Vehicle Data Summary:")
print(summarize_data(vehicle_data))

# Plotting the data
plot_data(cycle_data, 'Cycle Traffic Over Time')
plot_data(pedestrian_data, 'Pedestrian Traffic Over Time')
plot_data(old_pedestrian_data, 'Old Pedestrian Traffic Over Time')
plot_data(vehicle_data, 'Vehicle Traffic Over Time')
```
